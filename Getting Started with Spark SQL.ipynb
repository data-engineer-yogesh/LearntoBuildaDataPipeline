Why This Happens

dbfs:/databricks-datasets/... is legacy DBFS storage.

When you're using Unity Catalog, Databricks enforces stricter governance rules, and DBFS is not allowed as a data source for external tables.

Thatâ€™s why you see:
"Creating table in Unity Catalog with file scheme dbfs is not supported.
SQLSTATE: 0AKUC"

Solutions  we use 


```sql
CREATE TABLE default.online_retails
AS
SELECT *
FROM read_files(
  'dbfs:/databricks-datasets/online_retail/data-001/data.csv',
  format => 'csv',
  header => 'true',
  inferSchema => 'true'
);
```
